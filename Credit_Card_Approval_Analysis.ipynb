```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df["GENDER"]=df["GENDER"].apply(lambda x:1 if x=="F" else 0)
df.drop(columns=["Type_Income","EDUCATION","Marital_status", "Housing_type","Type_Occupation" ],inplace= True)
```

```python
target=pd.read_csv("Credit_card_label.csv")
```

```python
target.drop(columns=["Ind_ID"], inplace=True)
```

```python
X=df
```

```python
X
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,target, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random=RandomForestClassifier()
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
neighbors=KNeighborsClassifier()
logistic= LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def modelPerformance(predictions):
    print("accuracy score of model is {}".format(accuracy_score(y_test,predictions)))
```

```python
logistic.fit(X_train,y_train)
```

```python
logPredict=logistic.predict(x_test)
modelPerformance(logPredict)
```

```python
random.fit(X_train,y_train)
```

```python
randomPredict=random.predict(x_test)
modelPerformance(randomPredict)
```

```python
neighbors.fit(X_train,y_train)
```

```python
neighborsPredict=neighbors.predict(x_test)
modelPerformance(neighborsPredict)
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5]
}

grid_search = GridSearchCV(random, param_grid, cv=3)
grid_search.fit(X_train, y_train)

```

```python
print("Best parameters:", grid_search.best_params_)
```

```python
rfcPredict=grid_search.predict(x_test)
modelPerformance(rfcPredict)
```

```python
param_grid={
    "n_neighbors":[3,5,7,10],
    "weights": ["uniform","distane"],
    "metric": ["euclidean", "manhattan"]
     
}
gridNeighbors=GridSearchCV(neighbors,param_grid, cv=5)
gridNeighbors.fit(X_train,y_train)
print("best paarmetres", gridNeighbors.best_params_)
```

```python
gridNeighborPredict=gridNeighbors.predict(x_test)
modelPerformance(gridNeighborPredict)
```

```python
param_grid={
    "C":[0.01,0.1,1,10],
    "penalty":['l1','l2'],
    "solver":["liblinear","saga"]
}
gridLogistic=GridSearchCV(logistic,param_grid, cv=3)
gridLogistic.fit(X_train,y_train)
```

```python
print("best parametres", gridLogistic.best_params_)
```

```python
gridLogisticPredict=gridLogistic.predict(x_test)
modelPerformance(gridLogisticPredict)
```
